```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Points of the fish
X = np.loadtxt('fish.txt')
X_t = X.T
plt.plot(X_t[0,:], X_t[1,:], "bo")
```

## Transformation Matricies

### Scaling
\begin{align}
	S =
	\begin{bmatrix}
		\dfrac{1}{4} & 0 \\
		0 & \dfrac{1}{4}
	\end{bmatrix}.
\end{align}


### Rotation

$$\theta = 50$$

\begin{align}
	R =
	\begin{bmatrix}
		\cos\theta & -\sin\theta \\
		\sin\theta &  \cos\theta
	\end{bmatrix}.
\end{align}


### Horizontal shear

\begin{align}
	H =
	\begin{bmatrix}
		1 &  1 \\
		0 &  1
	\end{bmatrix}
\end{align}

```python
# Scaling Matrix
S = np.array([[0.25, 0],
              [0, 0.25]])

# Rotation Matrix
theta = 50
R = np.array([[math.cos(theta), -math.sin(theta)],
              [math.sin(theta), math.cos(theta)]])

# Horizontal Shear Matrix
H = np.array([[1, 1],
              [0, 1]])
```

```python
Y_s = S @ X_t

plt.plot(X_t[0,:], X_t[1,:], "bo", label="Original") # Plot original points in blue
plt.plot(Y_s[0,:], Y_s[1,:], "ro", label="Scaled") # Plot new points in red
plt.legend()
```

```python
# Sheared
Y_h = H @ X_t

plt.plot(X_t[0,:], X_t[1,:], "bo", label="Original") # Plot original points in blue
plt.plot(Y_h[0,:], Y_h[1,:], "ro", label="Sheared") # Plot new points in red
plt.legend()
```

```python
# Sheared
Y_r = R @ X_t

plt.plot(X_t[0,:], X_t[1,:], "bo", label="Original") # Plot original points in blue
plt.plot(Y_r[0,:], Y_r[1,:], "ro", label="Rotated") # Plot new points in red
plt.legend()
```

```python
# Creates new array with multiplied matricies
Y = S @ H @ R @ X_t

plt.plot(X_t[0,:], X_t[1,:], "bo", label="Original") # Plot original points in blue
plt.plot(Y[0,:], Y[1,:], "ro", label="Transformed") # Plot new points in red
plt.legend()
```
